```python
# One-hot encoding
x_train_channel_onehot = tf.keras.utils.to_categorical(x_train_channel)
x_test_channel_onehot = tf.keras.utils.to_categorical(x_test_channel)

print(x_train_channel_onehot[0].shape)
print(x_train_channel_onehot[0])
print(len(x_train_channel_onehot))
print(np.shape(x_train_channel_onehot))
# print(np.shape(x_test_channel_onehot))
```

```python
k = 5 # split the data into 5 buckets; 1 bucket = test set
kf = KFold(n_splits=k, shuffle=True)

r2_score_res = []
fold_no = 1
for train, test in kf.split(x_train_channel_onehot, x_train_channel_onehot):
    
    # details changed/omitted
    inputs = keras.Input(shape=(1,50))
    re = layers.Reshape((1,1,50))(inputs)
    dense_1 = layers.[layer info omitted](filters=1, activation="[omitted]", padding='same')
    output_1 = dense_1(re)
    x = layers.Flatten()(output_1)
    x = layers.Dense(1, kernel_regularizer='[omitted]')(x)
    Dropout(0.1)(x)
    outputs = Dense(1, activation='relu')(x)

    model = keras.Model(inputs=inputs, outputs=outputs, name="model")

    # model.summary()

    keras.utils.plot_model(model, "model.png", show_shapes=True)

    model.compile(loss='[omitted]', optimizer='[omitted]', metrics=['[omitted]'])
    
    # Generate a print
    print('------------------------------------------------------------------------')
    print(f'Training for fold {fold_no} ...')
    
    x_train_query, x_test_query = x_train_channel_onehot[train], x_train_channel_onehot[test]
    y_train_query, y_test_query = y_train_final[train], y_train_final[test]

    history = model.fit(x_train_query, y_train_query, epochs=5, batch_size=5)
    
    test_scores = model.evaluate(x_test_query, y_test_query, verbose=2)
    # print("Test loss:", test_scores[0])
    # print("Test accuracy:", test_scores[1])
    
    pred_values = model.predict(x_test_query)
    # print(y_test_query)
    # print(pred_values.squeeze())

    acc = r2_score(y_test_query, pred_values)
    # print(acc)
    
    fold_no += 1
    
print("End of training")
```
